```python

import numpy as np
np_array=np.array([1,2,3])

print("For 1 - D , shape gives number of element")
print(np_array)
print(np_array.shape)
print(np_array.ndim)

b = np.array([(1,2,3,4),(5,6,7,8)])
print("For 2 - D , shape gives rows,col")
print(b)
print(b.shape)
print(b.ndim)
```

```python
#ZEROES,ONES,FULL

x = np.zeros((4,5))
print(x)
x = np.ones((4,5))
print(x)
z = np.full((5,4),5)
print(z)
a = np.eye(5)
print(a)

# create a numpy array with random values
b = np.random.random((3,4))
print(b)
```

```python
#Linspace to get specific number of elemnts in specific range
d = np.linspace(10,30,5)
print(d)

# array of evenly spaced values --> specifying the step
e = np.arange(10,30,5)
print(e)
```

```python
#Making Random Array with shape & range being specified
c = np.random.randint(10,90,(5,5))
print(c)
print(c.shape)
print(c.size)
print(c.ndim)
```

```python
array=np.random.randint(10,20,(3,4))
print(array)
print(array.shape)
print("------The Transpose-----")
trans = np.transpose(array)
print(trans)
print(trans.shape)


b = array.reshape(6,2)
print(b)
print(b.shape)
```
